```python

import matplotlib.pyplot as plt
import numpy as np

# List of weights
list_1 = [48.49, 67.54, 57.47, 68.17, 51.18, 68.31, 50.33, 66.7, 45.62, 43.59, 
          53.64, 70.08, 47.69, 61.27, 44.14, 51.62, 48.72, 65.11]

# Compute the histogram
counts, bins = np.histogram(list_1, bins=5)

# Find the index of the bin with the maximum entries
max_count_index = np.argmax(counts)

# Get the range of the bin with the maximum entries
max_bin_range = (bins[max_count_index], bins[max_count_index + 1])

# Plotting the histogram
plt.hist(list_1, bins=5, edgecolor='black')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.title('Histogram of Weights')

# Highlight the bin with the maximum entries
plt.axvspan(max_bin_range[0], max_bin_range[1], color='red', alpha=0.3)

# Display the plot
plt.show()

# Print the range of the bin with the maximum entries
print(f'The bin with the maximum entries is in the range: {max_bin_range}')
```

```python
# box plot
import matplotlib.pyplot as plt
import numpy as np

#Find the interquartile range for the box plot created over the following list:

list_1 = [48.49, 67.54, 57.47, 68.17, 51.18, 68.31, 50.33, 66.7, 45.62, 43.59, 53.64, 70.08, 47.69, 61.27, 44.14, 51.62, 48.72, 65.11]
np_list = np.array(list_1)
percntile = [25,75]
box_percntile = np.percentile( np_list,percntile)
print(box_percntile) #The IQR can be calculated by finding the 25th and 75th percentile values.
```
