Q4:Plot a histogram for the Mins column with 15 bins. Among the three ranges mentioned below, which one has the highest frequency?

A - [54.6,68)

B - [68,81.4)

C - [121.6,135)

    a)A - [54.6,68)
    b)B - [68,81.4)

    d)All the bin ranges have the same frequency


```python
#Your code here

df.Mins.isnull().sum()

df['Mins'] = df['Mins'].replace(r'[-,]', 0, regex=True)
df['Mins'] = pd.to_numeric(df['Mins'], errors='raise', downcast=None)
print(df.Mins.value_counts())
```
